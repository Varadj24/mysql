```python
#Q1 What is a database? Differentiate between SQL and NoSQL databases.
```

```python
'''1)SQL (Relational Databases):

A)Data Structure: SQL databases use a tabular structure with rows and columns to store data. 
Data is organized into tables with predefined schemas, and each table is related to others
through keys (e.g., primary keys and foreign keys).

B)Data Integrity: SQL databases enforce data integrity constraints, 
such as primary keys, unique constraints, and foreign keys, 
to ensure data consistency and prevent data anomalies.

C)Query Language: SQL databases use the Structured Query Language (SQL) for querying and manipulating data. 
SQL provides a powerful and standardized way to retrieve and manage data.

D)Scaling: SQL databases are typically vertically scalable, 
meaning you can increase their capacity by upgrading hardware resources (e.g., CPU, RAM) of a single server. 
They may be less suitable for massive horizontal scaling.

E)Use Cases: SQL databases are commonly used for applications where data consistency, 
structured schemas, and ACID (Atomicity, Consistency, Isolation, Durability) transactions are essential. 
Examples include financial systems, e-commerce platforms, and content management systems.


2)NoSQL (Non-Relational Databases):

A)Data Structure: NoSQL databases use various data models, including document-oriented, key-value, 
column-family, and graph-based. These models allow for flexible and schema-less data storage.

B)Data Integrity: NoSQL databases often provide eventual consistency rather than strong consistency. 
They may relax some of the traditional ACID properties in favor of scalability and performance.

C)Query Language: NoSQL databases may use query languages specific to their data model. 
Some may support SQL-like queries, while others use different query languages or APIs.

D)Scaling: NoSQL databases are generally designed for horizontal scalability, 
making them suitable for handling large volumes of data and high-velocity workloads across distributed clusters of servers.

E)Use Cases: NoSQL databases are well-suited for applications with dynamic and evolving data requirements, 
high volumes of data, and distributed architectures. Examples include social media platforms, 
real-time analytics systems, and IoT (Internet of Things) data storage.'''
```

```python
#Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.
```

```python
'''DDL stands for "Data Definition Language." It is a subset of SQL (Structured Query Language) used for 
defining and managing the structure or schema of a database

A)CREATE:
Usage: The CREATE statement is used to create new database objects, such as tables, indexes, views, and constraints.

Example: Creating a new table named "Customers" with columns for customer information:

CREATE TABLE Customers (
  CustomerID INT PRIMARY KEY,
  FirstName VARCHAR(50),
  LastName VARCHAR(50),
  Email VARCHAR(100)
);

B)DROP:
Usage: The DROP statement is used to delete existing database objects, such as tables, indexes, views, and constraints. 
It permanently removes the object from the database.

Example: Dropping the "Customers" table:


DROP TABLE Customers;


C)ALTER:
Usage: The ALTER statement is used to modify the structure of an existing database object,
such as adding, modifying, or dropping columns from a table.

Example: Adding a new column "PhoneNumber" to the "Customers" table:

ALTER TABLE Customers
ADD PhoneNumber VARCHAR(20);


D)TRUNCATE:
Usage: The TRUNCATE statement is used to remove all rows from a table while retaining the table's structure. 
It is faster and uses fewer system resources than the DELETE statement for removing all data from a table.

Example: Truncating all data from the "Customers" table:

TRUNCATE TABLE Customers;'''
```

```python
#Q3. What is DML? Explain INSERT, UPDATE, and DELETE wit an example.
```

```python
'''DML stands for "Data Manipulation Language." It is a subset of SQL (Structured Query Language) used for manipulating data stored in a database.
DML statements are responsible for inserting, updating, and deleting data within database tables

1)INSERT:

Usage: The INSERT statement is used to add new rows (records) of data into a table.
Example: Inserting a new customer record into the "Customers" table:

INSERT INTO Customers (CustomerID, FirstName, LastName, Email)
VALUES (1, 'John', 'Doe', 'john.doe@example.com');

2)UPDATE:

Usage: The UPDATE statement is used to modify existing data in a table. It allows you to 
change values in one or more rows based on a specified condition.
Example: Updating the email address for a specific customer in the "Customers" table:

UPDATE Customers
SET Email = 'new.email@example.com'
WHERE CustomerID = 1;

3)DELETE:

Usage: The DELETE statement is used to remove rows from a table based on a specified condition.
Example: Deleting a customer record from the "Customers" table:

DELETE FROM Customers
WHERE CustomerID = 1;'''
```

```python
#Q5. Explain Primary Key and Foreign Key.
```

```python
'''1).Primary Key:

1)Definition: 
   A primary key is a constraint in a relational database table that uniquely identifies each record (row) in the table. 
It ensures that no two records in the table have the same value for the primary key column(s).

2)Properties:

A)Uniqueness: Each value in the primary key column(s) must be unique across all rows in the table.

B)Non-null: The primary key column(s) cannot contain NULL (missing) values.

C)Unchanging: Primary key values should generally not change once assigned to a record.

Purpose: The primary key serves as a means to identify and reference individual records within the table. 
It establishes a relationship between the data in the table and other tables in the database.

3)Example: 
   In a "Customers" table, the "CustomerID" column can be designated as the primary key. It ensures that each customer has a unique identifier.
   
2).Foreign Key:

1)Definition: A foreign key is a constraint in a relational database table that establishes 
a link between two tables based on the values of one or more columns. 
The foreign key in one table refers to the primary key in another table.

2)Properties:
A)Referential Integrity: The foreign key enforces referential integrity, 
ensuring that values in the foreign key column(s) match values in the primary key column(s) of the referenced table.

B)Relationship: Foreign keys create relationships between tables, allowing data to be linked and ensuring data consistency.

3)Purpose: Foreign keys are used to create associations and relationships between tables in a database. 
They enable the enforcement of data integrity rules and support the concept of data normalization.

4)Example: In an "Orders" table, the "CustomerID" column can be a foreign key that references 
the "CustomerID" column in the "Customers" table. This establishes a relationship between orders and customers,
ensuring that orders are associated with valid customers.'''
```

```python
#Q7. Give the order of execution of SQL clauses in an SQL query.
```

```python
'''
 The order of execution of SQL clauses in a standard SQL query is as follows:

1).SELECT: 
The SELECT clause is the first to be executed. It specifies which columns should be included in the result set and 
may include expressions and functions to transform or aggregate data.

FROM: 
The FROM clause is executed after the SELECT clause. It specifies the tables from which the data should be retrieved.
If multiple tables are involved (e.g., in a JOIN operation), this clause defines the source tables.

WHERE: 
The WHERE clause is executed after the FROM clause. It filters the rows from the tables specified in the
FROM clause based on a specified condition. Rows that meet the condition are included in the result set.

GROUP BY: 
If a GROUP BY clause is present, it is executed after the WHERE clause. 
It groups rows with similar values in specified columns into summary rows.
Typically, aggregate functions like SUM, COUNT, AVG, etc., are used in conjunction with GROUP BY.

HAVING: 
The HAVING clause is executed after the GROUP BY clause. It filters the grouped rows based on a condition, 
similar to the WHERE clause but applied to grouped rows.

ORDER BY: The ORDER BY clause is executed after all previous clauses. It sorts the result set based on specified columns,
either in ascending (ASC) or descending (DESC) order.

LIMIT/OFFSET: 
If a LIMIT and/or OFFSET clause is specified, they are executed last.
LIMIT restricts the number of rows returned in the result set, while OFFSET skips a specified number of rows before starting to return rows.
```
